```python
from nltk.stem import PorterStemmer
from collections import Counter
# Instantiate a PorterStemmer
stemmer = PorterStemmer()
```

```python
# Mapping list names to the corresponding tag
tag_dict = {'A1': 7, 'A2': 8, 'B1': 9, 'B2': 10, 'C1': 11}

# Iterate over the rows in the dataframe
for idx, row in train.iterrows():
    tokens = row['tokens']
    tags = row['ner_tags']

    # Check if the tag related to a word is 0
    for i in range(len(tags)):
        if tags[i] == 0:
            # Apply PorterStemmer on the token
            stemmed_token = stemmer.stem(tokens[i])

            # Check if the stemmed word lies in A1, A2, B1, B2 or C1
            for list_name, word_list in difficulty.items():
                if stemmed_token in word_list:
                    # Replace the 0 with 10, 11, 12, 13 or 14 respectively
                    tags[i] = tag_dict[list_name]

    # Update the row in the dataframe
    train.at[idx, 'ner_tags'] = tags
```

```python
# Create a dictionary with the counts of all tokens with tag 0
zero_tokens = [token for idx, row in train.iterrows() for i, token in enumerate(row['tokens']) if row['ner_tags'][i] == 0]

token_counts = Counter(zero_tokens)

num_rows = len(train)

# Iterate over the rows in the dataframe
for idx, row in train.iterrows():
    tokens = row['tokens']
    tags = row['ner_tags']

    # Check if the tag related to a word is 0
    for i in range(len(tags)):
        if tags[i] == 0:
            # Apply PorterStemmer on the token
            stemmed_token = stemmer.stem(tokens[i])

            # Check if the stemmed word lies in A1, A2, B1, B2 or C1
            for list_name, word_list in difficulty.items():
                if stemmed_token in word_list:
                    # Replace the 0 with 10, 11, 12, 13 or 14 respectively
                    tags[i] = tag_dict[list_name]
                    break  # Skip to next token once a match is found
            else:
                # If token was not found in any list, apply the count rules
                token_count = token_counts[tokens[i]]
                if token_count > num_rows * 0.9:
                    tags[i] = 7
                elif token_count > num_rows * 0.7:
                    tags[i] = 8
                elif token_count > num_rows * 0.5:
                    tags[i] = 9
                elif token_count > num_rows * 0.3:
                    tags[i] = 10
                elif token_count < num_rows * 0.3:
                    tags[i] = 11

    # Update the row in the dataframe
    train.at[idx, 'ner_tags'] = tags
```

```python
def tokenize_adjust_labels(all_samples_per_split):
  tokenized_samples = tokenizer.batch_encode_plus(all_samples_per_split["tokens"], is_split_into_words=True)

  total_adjusted_labels = []
  print(len(tokenized_samples["input_ids"]))

  for k in range(0, len(tokenized_samples["input_ids"])):
    prev_wid = -1
    word_ids_list = tokenized_samples.word_ids(batch_index=k)
    existing_label_ids = all_samples_per_split["ner_tags"][k]
    i = -1
    adjusted_label_ids = []

    for wid in word_ids_list:
      if(wid is None):
        adjusted_label_ids.append(-100)
      elif(wid!=prev_wid):
        i = i + 1
        adjusted_label_ids.append(existing_label_ids[i])
        prev_wid = wid
      else:
        adjusted_label_ids.append(existing_label_ids[i])

    total_adjusted_labels.append(adjusted_label_ids)

  tokenized_samples["labels"] = total_adjusted_labels
  return tokenized_samples
```

```python
tokenized_dataset = train.map(tokenize_adjust_labels, batched=True, remove_columns=['tokens', 'ner_tags'])
```

```python
from transformers import AutoModelForTokenClassification, TrainingArguments, Trainer
import numpy as np
from datasets import load_metric
metric = load_metric("seqeval")
def compute_metrics(p):
    predictions, labels = p
    predictions = np.argmax(predictions, axis=2)

    # Remove ignored index (special tokens)
    true_predictions = [
        [p for (p, l) in zip(prediction, label) if l != -100]
        for prediction, label in zip(predictions, labels)
    ]
    true_labels = [
        [l for (p, l) in zip(prediction, label) if l != -100]
        for prediction, label in zip(predictions, labels)
    ]

    results = metric.compute(predictions=true_predictions, references=true_labels)
    flattened_results = {
        "overall_precision": results["overall_precision"],
        "overall_recall": results["overall_recall"],
        "overall_f1": results["overall_f1"],
        "overall_accuracy": results["overall_accuracy"],
    }
    for k in results.keys():
        if(k not in flattened_results.keys()):
            flattened_results[k+"_f1"]=results[k]["f1"]

    return flattened_results
```

```python
def predict(input_text):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

    # Send model to device
    model.to(device)

    # Tokenize the input text
    tokenized_input = tokenizer.encode_plus(input_text, truncation=True, padding=True, return_tensors="pt")

    # Remove 'token_type_ids' from the inputs
    tokenized_input.pop('token_type_ids', None)

    # Send tokenized inputs to device
    tokenized_input = {k: v.to(device) for k, v in tokenized_input.items()}

    # Pass the input through the model
    output = model(**tokenized_input)

    # Get the predicted tags (the argmax of the logits)
    predicted_tag_ids = np.argmax(output.logits.detach().cpu().numpy(), axis=2)

    return predicted_tag_ids
```
